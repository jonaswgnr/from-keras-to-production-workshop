# Images & Matplotlib

- Create a list of file names in the validation directory and
- print randomly 16 images:

```python
valid_image_files_path = "/keras2production/fruits/Test/"

pictures = !find $valid_image_files_path -type f -name "*.jpg"
samplePictures = random.sample(pictures, 16)

figure = plt.figure(figsize=(12,12))
i=1
for picture in samplePictures:
    figure.add_subplot(6,6,i)
    plt.axis('off')
    img = plt.imread(picture)
    plt.imshow(img)
    i=i+1 
```
