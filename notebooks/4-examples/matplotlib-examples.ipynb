```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Matplotlib

We will often create different kinds of visualiziation during this workshop, such as plotting data or images.<br>
*matplotlib* is the most popular Python library suitable for this task. It enables us producing plots and other visualizations.

## Basic plots

Everthing in matplotlib is part of a *figure*. A figure can contain multiple plotting spaces, called *axes*. Plots like *line plots* live inside an axes.

```python
fig, axes = plt.subplots(1,2)
fig.set_size_inches(12, 6)
fig.suptitle("Example Plot")

axes[0].set_title("Linear")
axes[0].set_xlabel("Number")
axes[0].set_ylabel("Value")
axes[0].set_xticks([0, 25, 50, 75, 100])
axes[0].plot(np.arange(100), np.arange(100))

axes[1].set_title("Sine")
axes[1].set_xlabel("Sine(Number)")
axes[1].set_ylabel("Value")
axes[1].set_axis_off()
axes[1].plot(np.arange(100), np.sin(np.arange(100)), 'b--')

fig.tight_layout(pad=5, w_pad=5, h_pad=1.0)
plt.show()
```

## Histograms

An axes can also contain multiple plots - like a *histogram* combined with a *line plot*.

```python
# The data that we want to plot
mu = 0
sigma = 0.1
distribution = np.random.normal(mu, sigma, 1000)

# Creating a figure with a single plotting space (axes)
fig, axes = plt.subplots(1,1)
fig.set_size_inches(8,6)
fig.suptitle("Combined histogram and line plot")

# Create the histogram
_, bins, _ = axes.hist(distribution, bins=30, density=True)
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

# Plot
axes.plot(bins, y, 'r+-')
plt.show()
```

## Scatterplot

For less complex plots, e.g. figures with just a single axes, there is no need for a subplot as pyplot can be used directly.

```python
m = np.random.rand(3, 100) * 100
plt.scatter(x=m[0], y=m[1], c=m[2])
plt.show()
```

## Images

```python
m = np.random.rand(100, 100, 3)
plt.imshow(m)
print(m.shape)
```

Even multiple subplots can be generated on the fly using pyplot.

```python
rows = 5
cols = 5
number_of_images = rows * cols

for index in range(number_of_images):
    plt.subplot(rows, cols, index+1)
    img = np.random.rand(100, 100, 3)
    plt.imshow(img)
    plt.xticks([]), plt.yticks([])
```
